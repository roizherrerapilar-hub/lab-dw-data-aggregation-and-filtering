```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.shape
```

```python
df.info() #para tener visibilidad del nombre de las columnas. 
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')   #vamos a corregir el nombre de las columnas para empezar con el analisis. 
df.head()
```

```python
#1. Crear un dataframe con clientes que: tiene pocas reclamaciones (inferior a 1000) y han respondido Si a la campaña. Lo que tenemos que hacer es filtrar los datos. 

new_dataframe = df[(df['total_claim_amount'] < 1000) & (df['response'] == 'Yes')]
new_dataframe
```

```python
#2. Calcular medias de monthly_premium y de customer_lifetime_value agrupado por: policy_type y gender considerando la response: yes y trabajando con el original dataframe y compararlo con total_claim_amount

yes_response = df[df['response'] == 'Yes']
yes_response
```

```python
customer_lifetime_mean = yes_response.groupby(['policy_type', 'gender'])['customer_lifetime_value'].mean()
customer_lifetime_mean
```

```python
monthly_premium_auto_mean = yes_response.groupby(['policy_type', 'gender'])['monthly_premium_auto'].mean()
monthly_premium_auto_mean
```

```python
total_claim_amount_mean = yes_response.groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
total_claim_amount_mean
```

```python
mean_comparison = pd.concat([customer_lifetime_mean, monthly_premium_auto_mean, total_claim_amount_mean], axis=1)
mean_comparison.unstack()
```

Insights del analisis: 
- Las mujeres con pólizas Personal Auto son el segmento más rentable para la empresa.
- Hombres con pólizas Special Auto también representan clientes de alto valor.
- Personal auto policy es el grupo más arriesgado, ya que son los que generan más reclamaciones. 
- Vemos que el grupo de personal auto es muy rentable para la empresa pero con un riesgo a considerar. 

```python
#3. Analizar cuantos clientes hay por state que tienen poliza. Y filtra el resultado incluyendo solo aquellos states con mas de 500 paises. 

state_customers = df.groupby('state')['customer'].count()
state_customers
```

```python
filtered_states = state_customers[state_customers > 500]
filtered_states
```

```python
#4. Calcula el max, min, mediana de customer lifetime value por education level y gender. 

customer_lifetime_max = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg('max')
customer_lifetime_max
```

```python
customer_lifetime_min = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg('min')
customer_lifetime_min
```

```python
customer_lifetime_median = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg('median')
customer_lifetime_median
```

```python
comparison_table = pd.concat([customer_lifetime_max, customer_lifetime_min, customer_lifetime_median], axis=1)
comparison_table.unstack().round(2)
```

Insights:
- Hay una gran diferencia entre el valor máximo y la mediana, lo que indica pocos clientes altamente valiosos. 
- Los clientes con menor nivel educativo pueden alcanzar valores de vida muy altos, lo que rompe la idea de que mayor educación implica mayor valor.
- El género no tiene un impacto consistente en el valor del cliente, ya que su efecto varía según el nivel educativo.
